```python
website = 'https://www.gutenberg.org/cache/epub/47287/pg47287-images.html'
result = requests.get(website)
content = result.text

soup = BeautifulSoup(content, 'lxml')
print(soup.prettify())
```

image.png

```python
title = soup.find(id = "HISTORIA_DEL_REY_SCHAHRIAR_Y_DE_SU_HERMANO_El_REY_SCHAHZAMAN")
print(title)
```

```python
titulo = soup.find(id="HISTORIA_DEL_MERCADER_Y_EL_EFRIT").get_text()
print(titulo)
```

```python
boxes = soup.findAll('div', class_='figcenter')
for box in boxes:
    box2 = box.find('h2')
    if box2 != None:
        print(box2.get_text())
```

```python
boxes = soup.findAll('h2')
for box in boxes:
    if box != None:
        title = box.get_text()
        if title.upper()[:8] =="HISTORIA" :
            print(title)
```
